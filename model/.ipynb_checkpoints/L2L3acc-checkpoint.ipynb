```python
from sklearn.metrics import adjusted_rand_score
from matplotlib import pyplot as plt
from scipy.special import softmax
from itertools import product
import numpy as np
from utils import *
from model import *
import time
import seaborn as sns
sns.set_context('talk')

%load_ext autoreload
%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
# model 13
mse_top_k = 3
schargs = {
   'concentration':SCHARGS_MSE[mse_top_k]['concentration'],
   'stickiness_wi':SCHARGS_MSE[mse_top_k]['stickiness_wi'],
   'stickiness_bt':SCHARGS_MSE[mse_top_k]['stickiness_bt'],
   'sparsity':SCHARGS_MSE[mse_top_k]['sparsity'],
   'pvar': 0,
   'lrate':1,
   'lratep':1,
   'decay_rate':1,
}
semargs = {
  'beta2':False,
  'skipt1':True,
  'ppd_allsch':False
}
```

```python
taskargs = {
  'condition':'blocked',
  'n_train':160,
  'n_test':40
}
args = {
    'sem':semargs,
    'sch':schargs,
    'exp':taskargs
}
param_str = "-".join(["%s_%.3f"%(i,j) for i,j in schargs.items()])
param_str += "-"+"-".join(["%s_%.3f"%(i,j) for i,j in semargs.items()])
param_str
```

```python
## timestamp and dir for saving
import os
tstamp = time.perf_counter_ns()
figdir = 'figures/single_model_nb/L2L3acc-%s-%i/'%(param_str,tstamp)
os.makedirs(figdir)
```

### main

```python
ns = 50
L = []
condL = ['blocked','interleaved',
         'early','middle','late'
        ]
exp_batch_data = run_batch_exp_curr(ns,args,condL) # [curr,seeds,{data}]
```

```python
## accuracy
# bads = get_bad_seed(exp_batch_data,max_splits=50)
# goods = np.setdiff1d(np.arange(ns),bads)
batch_acc = unpack_acc(exp_batch_data,False) # curr,seeds,tsteps,trials
```

```python
print(batch_acc.shape)
mean_acc = batch_acc.mean(1) # over seeds
test_acc = mean_acc[:,-40:].mean(1) # curr  
mean_acc.shape # cond,step,trial
```

```python
f,axar=plt.subplots(2,1,figsize=(20,20),sharex=True)
for cidx in range(len(condL)):
  for tidx in range(2):
    ax = axar[tidx]
    ax.plot(mean_acc[cidx,tidx,:],label=condL[cidx])

    ax.axhline(0.5,c='k',ls='--')
    ax.set_ylim(-0.05,1.01)

    ax.grid(True,axis='y')

plt.legend()
plt.title(param_str)
plt.savefig(figdir+'acc-by-time-Split.png')
```
