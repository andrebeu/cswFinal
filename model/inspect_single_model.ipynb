# Inspect single model
* this script can be run to inspect new model params, or generate the full simulation dataset for pre-selected model params (e.g. top 5 MSE models)
* use this script to:
    * generate full simulation dataset from a specified model parametrization

### adjusted rand score

```python
def calc_adjrand(exp_batch_data):
  arscores = -np.ones([len(condL),ns,3])
  for curr_idx in range(len(condL)):
    for seed_idx in range(ns):
      for t_idx,tstep in enumerate([0,2,3]):
        arscores[curr_idx,seed_idx,t_idx] = adjusted_rand_score(
          exp_batch_data[curr_idx][seed_idx]['exp'][:,1],
          exp_batch_data[curr_idx][seed_idx]['zt'][:,tstep]
        )
  return arscores

arscores = calc_adjrand(exp_batch_data)
arscores.shape # cond, seed, tstep
```

```python
f,axar=plt.subplots(1,3,figsize=(30,6),sharey=True)
for t in range(3):
  ax=axar[t]
  ax.violinplot(arscores[:,:,t].T,showextrema=1,showmeans=1)
  for c in range(len(condL)):
    ax.scatter(np.repeat(c+1,ns),arscores[c,:,t].T)
  ax.set_xticks(np.arange(1,len(condL)+1))
  ax.set_xticklabels(condL)
  ax.set_title(['0','2','3'][t])
plt.suptitle(param_str)
plt.savefig(figdir+'adjrand')
```

```python
def calc_adjrand_bytime(exp_batch_data,curr_idx=0,window=200):
  """ time binned adjusted rand
  return arscores (seed,tstep,trial_window)
  - single curr
  """
  n_trial_window = 200 - window
  arscores = -np.ones([ns,3,n_trial_window])
  for seed_idx in range(ns):
    for t_idx,tstep in enumerate([0,2,3]):
      ## model and experiment latent causes
      expLC = exp_batch_data[curr_idx][seed_idx]['exp'][:,1]
      modelLC = exp_batch_data[curr_idx][seed_idx]['zt'][:,tstep]
      for tr_idx in range(n_trial_window):
        wExpLC = expLC[tr_idx:tr_idx+window]
        wModLC = modelLC[tr_idx:tr_idx+window]
        ar = adjusted_rand_score(wExpLC,wModLC)
        arscores[seed_idx,t_idx,tr_idx] = ar
  return arscores

# arscores_bytime = calc_adjrand_bytime(exp_batch_data,window=10)
# print(arscores_bytime.shape) # seeds, tstep, timebin
```

### adjusted rand windowed in time for each timestep 

```python
curr = 1 
arscore_window = 10
arscores_bytime = calc_adjrand_bytime(exp_batch_data,curr_idx=curr,window=arscore_window)
print(arscores_bytime.shape) # seeds, tstep, timebin
```

# posterior, likelihood, prior
### analysis of model state

```python
def plt_prdata(dtype='post',schL=[0,1],curr_idx=0,seed_idx=0):
  """ 
  plots probability data (post,prior,like) 
  of at most 3 timesteps of 4 schemas
  NB single seed
  """
  # load data
  prdata = unpack_data(exp_batch_data,
          dtype=dtype)[curr_idx][seed_idx]
  # setup plots
  f,axar = plt.subplots(3,4,figsize=(20,10),sharex=True,sharey=True)
  # loop over timesteps and schema
  for ti,tstep in enumerate([0,2,3]):
    for si,sch in enumerate(schL):
      axar[0,si].set_title('sch%i'%sch)
      ax = axar[ti,si]
      # plot
      ax.plot(prdata[:,tstep,sch])
      for k in np.arange(0,200,40):
        ax.axvline(k,c='k',lw=0.5)
    axar[ti,0].set_ylabel('tstep%i'%tstep)
  title = '%s-%s'%(dtype,param_str)
  plt.suptitle(title)
```
